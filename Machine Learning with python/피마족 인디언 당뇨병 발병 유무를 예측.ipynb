피마족 인디언 당뇨병 발병 유무를 예측하는 이진 분류 문제 (1:당뇨병, 0: 정상)

1. 데이터 준비하기 !

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense

# 랜덤 시드 고정시키기
np.random.seed(5)
```

```python
dataset= np.loadtxt('diabetes.csv', delimiter=',', skiprows=1)
dataset
#데이터 셋이 숫자라 넘파이로 불러올 수 있다.
#loadtxt()를 활용하여 처음부터 넘파이 배열로 불러올 수 있음.
```

데이터셋 생성하기 - 입력(속성값 8개feature :문제집), 출력 (판정결과 1개 :정답지) 변수로 분리



```python
# 학습 데이터 700개 and 8 변수 가져오기
X_train=dataset[:700, 0:8] #

y_train=dataset[:700, 8]

# 테스트 데이터 68개 and 1개 변수 가져오기
X_test=dataset[700:, 0:8]
y_test=dataset[700:, 8]
```

모델 구성하기

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu')) #첫번째 층은 8개 입력할거야, relu로 해줘
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

**모델 학습과정 설정하기**

```python
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy']) #리스트 형태로 되어있음
```

**모델 학습 시키기**

```python
history=model.fit(X_train, y_train,
                  epochs=1500,
                  batch_size=64)
```

**학습결과 그리기**

```python
import matplotlib.pyplot as plt

his_dict = history.history
loss = his_dict['loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize = (10, 5))

# 훈련 및 검증 손실 그리기
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color = 'orange', label = 'train_loss')
ax1.set_title('train loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc = his_dict['accuracy']

# 훈련 및 검증 정확도 그리기
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, acc, color = 'blue', label = 'train_accuracy')
ax2.set_title('train accuracy')
ax2.set_xlabel('epochs')
ax2.set_ylabel('accuracy')
ax2.legend()

plt.show()
```

모델 평가하기

```python
scores = model.evaluate(X_test, y_test)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

모델 성능 평가

```python
# ROC 커브

def plot_auc_roc(model):
    probs = model.predict_classes(X_test)
    fpr, tpr, threshold = metrics.roc_curve(y_test, probs)
    roc_auc = metrics.auc(fpr, tpr)

    # method I: plt
    import matplotlib.pyplot as plt
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```
