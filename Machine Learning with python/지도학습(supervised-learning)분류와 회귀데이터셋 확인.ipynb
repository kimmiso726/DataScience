예제에 사용할 데이터셋


1.   이진 분류 데이터셋(forge)확인하기

데이터 셋 다운로드

```python
import mglearn
import matplotlib.pyplot as plt

# 데이터셋 만들기
X,y = mglearn.datasets.make_forge()
```

데이터 확인하기

```python
print("X.shape", X.shape)
print("y.shape", y.shape)
```

```python
X
```

```python
plt.figure(dpi=200)
plt.rc('font',family='NanumBarunGothic')

#산점도 그리기
mglearn.discrete_scatter(X[:,0], X[:,1],y)

plt.legend(['클래스 0', '클래스 1'], loc=4)
plt.xlabel('첫 번째 특성')
plt.ylabel('두 번째 특성')
plt.show()
```

2. 회귀데이터 셋(wave) 확인하기

데이터 다운로드

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
```

데이터 확인하기

```python
print("X.shape:",X.shape)
print("y.shape:", y.shape)
```

```python
y
```

```python
# 산점도 그리기
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

plt.figure(dpi=100)
plt.rc('font',family='NanumBarunGothic')

plt.plot(X,y,'o')
plt.ylim(-3,3)
plt.xlabel('특성')
plt.ylabel('타깃')
plt.show()
```

분류 문제정의 : 위스콘신 유방암 데이터 셋을 사용한 악성 종양(Label,1) 예측하기

데이터 준비하기(https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic))


```python
from sklearn.datasets import  load_breast_cancer

cancer=load_breast_cancer()
print(cancer.keys()) # 데이터셋 키 확인
```

데이터 확인하기

```python
cancer.data.shape #유방암 데이터의 형태-> 569건, 특성30개
```

```python
# 클래스별 샘플(타깃) 개수

import numpy as np

# 양성, 악성 종양의 비율 37% : 63%
{n : v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}
```

```python
# 데이터의 특성 (feature) 확인
cancer.feature_names
```
