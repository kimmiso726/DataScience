```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv', encoding='cp949')
data= csv.reader(f)

m = []
f = []

for row in data :
  if '장항읍' in row[0] :
    for i in range(0,101) :
      m.append(-int(row[i+3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담음 -> 100세 이상 부터 값이 담김

f.reverse() #여자를 0세부터 담길 수 있도록 바꿔준다. 

print(m)
print(f)
print(len(m))
print(len(f))
```

데이터 시각화 하기

```python
import matplotlib.pyplot as plt
plt.figure(dpi=200)
plt.barh(range(101),m, label ='남성') # Make a horizontal bar plot
plt.barh(range(101),f, label='여성')
plt.legend()
plt.title('장항읍 지역의 남녀 성별 인구 분포')
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False #마이너스 기호 깨짐 방지
plt.show()

#남녀 성별 차이가 많지는 않은데 대체적으로 남자가 많고 
#젊은층은 적고 고령화가 심해진다.
#역시 여자가 오래산다.

```
