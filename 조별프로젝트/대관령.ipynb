1. 최고 기온과 최고기온 날짜 찾기


```python
import csv

max_temp = -999 #최고 기온 값을 저장할 변수, 명확한 결과를 위해 극적인 값으로 셋팅. 0으로 안함.
max_date = '' # 최고 기온이 가장 높았던 날짜를 저장하는 변수

f= open('/content/대관령.csv', 'r', encoding='cp949')  

data= csv.reader(f) 
header= next(data)  #헤더 추출

for row in data :
  #최고 기온 데이터가 '' 인지 확인 (결측치인지 확인)
  if row[-1] == '' :
    row[-1] =-999    #-999를 넣어서 빈 문자열('')이 있던 자리라고 표시
  
  row[-1] = float(row[-1]) #기온 데이터 실수형으로 변환

  #최고 기온 정보를 저장
  if max_temp < row[-1] :
    max_date = row[0]
    max_temp = row[-1]

f.close()

print('기상 관측 이래 대관령의 최고 기온이 가장 높았던 날은', max_date+'로', max_temp, '도 였습니다.')
```

**matplot생성**

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 임포트


f= open('대관령.csv', encoding='cp949')  

data= csv.reader(f) 
header= next(data)  #헤더 추출

result = []   #결측치를 제거한 최고 기온 리스트

for row in data :
  if row[-1] != '' :    #최고 기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1]))

result    

plt.figure(dpi=200)   #해상도
plt.plot(result, 'r')
plt.show()
```

```python
import csv

min_temp = 999 #누가봐도 최저기온 아닌 값 넣어줘. 최저 기온 값을 저장할 변수, 명확한 결과를 위해 극적인 값으로 셋팅. 0으로 안함.
min_date = '' # 최고 기온이 가장 높았던 날짜를 저장하는 변수

f= open('/content/대관령.csv', 'r', encoding='cp949')  

data= csv.reader(f) 
header= next(data)  #헤더 추출

row_list=[]

for row in data :
  if row[-1] == '' :
   row[-1] = -999  
  if row[3] == '' :
   row[3] = 999
  row[-1]=float(row[-1])
  row[3]=float(row[3])
  if row[-1] != -999 and row[3] != 999 :
    row_list.append(row[-1]-row[3])

print(row_list)

```
