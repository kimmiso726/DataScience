1) 히스토그램을 활용한 최고기온, 최저 기온, 평균기온 분포 확인하기


```python
#한글 깨짐 처리

import matplotlib as mpl
import matplotlib.pyplot as plt
 
%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
import matplotlib.font_manager as fm

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('/content/대관령.csv', encoding='cp949')
data= csv.reader(f)
next(data) #헤더

high = [] #최고기온
low = [] #최저기온
avg = []  #평균기온

for row in data :   #둘다 같은 결과여야 하기 때문에 
  if row[-1] != ''  and row[-2] != '' and row[-3] !='' :  #최고, 최저, 평균 결측치 제거
      high.append(float(row[-1]))
      avg.append(float(row[-3]))
      low.append(float(row[-2]))

#최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(high, bins=100, color='r', label='최고 기온') #구간옵션
plt.hist(low, bins=100, color='b', label='최저 기온')
plt.hist(avg, bins=100, color='g', label='평균 기온')
plt.title('대관령의 최고, 최저, 평균기온 그래프')
plt.legend()
plt.show()   
```

3) 월별 최고기온 상자 그림 그리기


```python
import csv
import matplotlib.pyplot as plt

f=open('/content/대관령.csv', encoding='cp949')
data= csv.reader(f)
next(data) #헤더

month = [] #월별 데이터를 저장할 리스트 month 변수 선언
for i in range(12) :
  month.append([])  #month 리스트 네에 12개 리스트 생성[[],[],[],[].....31개]

for row in data :
  if row[-1] != '' :
     month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(7,2),dpi=300)
plt.boxplot(month, showfliers=False)  #아웃라이어(이상치)값 생략
plt.show()           
```

4) 특정 달의 일별 최고 기온 상자 그림 그리기


```python
import csv
import matplotlib.pyplot as plt


f=open('/content/대관령.csv', encoding='cp949')
data= csv.reader(f)
next(data) #헤더

day = [] #일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31) :
  day.append([])  #day 리스트 네에 31개 리스트 생성[[],[],[],[].....31개]

for row in data :
  if row[-1] != '' :
    if row[0].split('-')[1] == '01' : # 1월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(20,2),dpi=300)
plt.boxplot(day, showfliers=False)  #아웃라이어(이상치)값 생략
plt.show()           
```
